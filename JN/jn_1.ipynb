```python
#saving images in grayscale and resizing them in a new folder
import cv2
import matplotlib.pyplot as plt
from pathlib import Path

# --- folders (edit these) ---
input_folder  = Path("dataset/raw/kambi")
output_folder = Path("dataset/processed/p_kambi/im_k")
output_folder.mkdir(parents=True, exist_ok=True)

valid_exts = {".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff"}

for img_path in input_folder.iterdir():
    if img_path.suffix.lower() in valid_exts:
        # read image with OpenCV
        img = cv2.imread(str(img_path))
        if img is None:
            print(f"Skipped: {img_path.name}")
            continue

        # convert to grayscale
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # resize to 256x256
        resized = cv2.resize(gray, (256, 256), interpolation=cv2.INTER_AREA)

        # save to output folder
        out_path = output_folder / img_path.name
        cv2.imwrite(str(out_path), resized)
        print(f"Saved {out_path.name}")
        
```
