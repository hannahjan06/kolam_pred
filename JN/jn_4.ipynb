```python
import numpy as np
import glob
import os

# Path to folder containing all individual npy files
input_dir = "dataset/processed/p_sikku/s_npy_224"
output_path = "dataset/processed/p_sikku/s_combined_kolam.npy"

# Get list of npy files
file_list = sorted(glob.glob(os.path.join(input_dir, "*.npy")))

all_rows = []
for f in file_list:
    img = np.load(f)                # shape: (H, W)
    flat = img.flatten()            # shape: (H*W,)
    all_rows.append(flat)

# Stack into a 2-D array: (num_images, H*W)
big_matrix = np.vstack(all_rows)

# Save single file
np.save(output_path, big_matrix)
print("Combined shape:", big_matrix.shape)
```
